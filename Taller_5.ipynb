```python
import numpy as np
import matplotlib.pyplot as plt

def calcular1(n):
  resultado = (1+1/n)**n
  return resultado

print("A")
for i in range(1,6):
  print(f"a{(i)} = {calcular1(i)}")
print("-----------------------------------------------")

print("B")
for i in range (1,201):
  print(f"a{(i)} = {calcular1(i)}")
print("-----------------------------------------------")

print("C")
n = np.arange(1,201)
an = calcular1(n)
e = np.exp(1)

plt.plot(n,an)
plt.axhline(y= e, color="red")
plt.show()
```

D) S√≠, porque se aproxima infinitamente a $e$ conforme $n$ crece, entonces es razonable afirmar que: $\lim_{n \to \infty } a_n = e$

```python
for i in range(1,101):
  if calcular1(i)>2.7:
    print(f"(E) El valor mas peque√±o para n para que ùëéùëõ > 2.7 es: {i}")
    break

for i in range (1,100001):
  resultado= abs(calcular1(i) - 2.71828)
  if resultado < 0.01:
    print(f"(F) El primer valor para n para que |ùëéùëõ ‚àí ùëí| < 0.01 es: {i} ")
    break
```

```python
def calcular2 (n):
  resultado = n * np.sin(1/n)
  return resultado
print("A")
for i in range(1,6):
  print(f"a{(i)} = {calcular2(i)}")
print("-----------------------------------------------")

print("B")
for i in range (1,201):
  print(f"a{(i)} = {calcular2(i)}")
print("-----------------------------------------------")

n = np.arange(1,201)
an = calcular2(n)
plt.plot(n,an)
plt.axhline(y =1, color="red")
plt.show()
```

D) S√≠, porque se aproxima infinitamente a $1$ conforme $n$ crece, entonces es razonable afirmar que: $\lim_{n \to \infty } a_n = 1$

```python
for i in range(1,10001):
  if calcular2(i) > 0.999:
    print(f"(E) El valor mas peque√±o para n para que ùëéùëõ > 0.999 es: {i}")
    break

for i in range(1,1000001):
  resultado= abs(calcular2(i) - 1)
  if resultado < 0.0001:
    print(f"(F) El primer valor para n para que |ùëéùëõ ‚àí 1| < 0.0001 es: {i}")
    break
```
